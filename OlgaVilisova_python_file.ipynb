```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv('source.csv')
```

```python
data['Datetime']=pd.to_datetime(data['Datetime'],utc=True)
```

```python
data.dtypes
```

```python
data.head(2)
```

```python
data['Datetime']=data['Datetime'].dt.tz_convert('Etc/GMT-6')
```

```python
data.head(2)
```

```python
data['total']=np.nan
```

```python
for row in range(len(data)):
    if data.iloc[row,0]=='ProductA' and data.iloc[row,4]=="Pure":
        data.iloc[row,-1]=data.iloc[row,2]*data.iloc[row,3]
    elif data.iloc[row,0]=='ProductB' and data.iloc[row,4]=="Pure":
        for line in range(len(data)):
            if data.iloc[line,0]=='ProductA' and data.iloc[row,1]==data.iloc[line,1]:
                data.iloc[row,-1]=(data.iloc[row,3]-data.iloc[line,3])*data.iloc[row,2]
    else: data.iloc[row,-1]=data.iloc[row,3]*0.75
```

```python
data.to_csv('result.csv',index=False)
```
